```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
#2 Load data
file_path = "/content/ML_project/gld_price_data.csv"
df = pd.read_csv(file_path)
```

```python
#9
#Two types : 1]Positive correlation 2] Negative correlation
#correlation = df.corr()

correlation = df.select_dtypes(include=['number']).corr()
```

```python
#10 constructing a heatmap to understand the correlatiom
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

```python
#11 correlation values of GLD
print(correlation['GLD'])
```

```python
#12 checking the distribution of the GLD Price
sns.distplot(df['GLD'],color='green')
```

```python
#13

from sklearn.model_selection import train_test_split

X = df.drop(['Date','GLD'], axis=1)
Y = df['GLD']

# here This line splits our dataset into two parts: one for training the model and one for testing it.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X)
```

```python
print(Y)
```

```python
#14
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#15
predictions = model.predict(X_test)
```

```python
#16 - R² score (coefficient of determination)
r2 = metrics.r2_score(Y_test, predictions)
print("R² Score:", r2)
```

```python
#17
# Mean Absolute Error
mae = metrics.mean_absolute_error(Y_test, predictions)
print("MAE:", mae)

# Mean Squared Error
mse = metrics.mean_squared_error(Y_test, predictions)
print("MSE:", mse)
```

```python
#18
plt.scatter(Y_test, predictions, color='blue')
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Actual vs Predicted Gold Price (GLD) - Linear Regression")

line = np.linspace(min(Y_test), max(Y_test), 100)
plt.plot(line, line, color='red', linestyle='--')

plt.show()
```
